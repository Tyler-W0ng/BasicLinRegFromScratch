```python
import numpy as np 
import matplotlib.pyplot as plt

x_train = np.array([1, 2, 2, 3, 5, 6, 1, 2, 9])
y_train = 4*x_train + np.array([1, 3, 4, 0, 10, 2, 5, 4, 6])

print(x_train, y_train)

w = 0
b = 0
learning_rate = .01
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
def compute_cost(x_train, y_train, w, b):
    m = x_train.shape[0]

    total_loss = 0
    for i in range(len(x_train)):
        f_wb = w * x_train[i] + b
        loss = (f_wb - y_train[i])**2
        total_loss += loss
    
    return .5 * (1/m) * total_loss
```

```python
cost = compute_cost(x_train, y_train, w, b)
print(cost)
```

```python
def compute_gradient(x_train, y_train, w, b):
    m = x_train.shape[0]
    grad_w = 0
    grad_b = 0

    for i in range(len(x_train)):
        f_wb = w * x_train[i] + b
        dj_dw = 2*(f_wb - y_train[i])*x_train[i]
        dj_db = 2*(f_wb - y_train[i])

        grad_w += dj_dw
        grad_b += dj_db

    return (1/m)*grad_w, (1/m)*grad_b
```

```python
grad_w, grad_b = compute_gradient(x_train, y_train, w, b)
print(grad_w, grad_b)
```

```python
def gradient_descent(x_train, y_train, grad_w, grad_b, w, b, learning_rate):
    m = x_train.shape[0]
    
    for epoch in range (50):

        w = w - learning_rate * grad_w
        b = b - learning_rate * grad_b

        grad_w, grad_b = compute_gradient(x_train, y_train, w, b)

        print(f"Epoch: {epoch}, w: {w}, b: {b}, cost: {compute_cost(x_train, y_train, w, b)}")

    return w, b
```

```python
w, b = gradient_descent(x_train, y_train, grad_w, grad_b, w, b, learning_rate)
```

```python
x_test = np.array([0, 2.4, 3.7, 8.1, 9.01, 1.4, 5.6, 2.7, 4.6, 7.8])

y_test = w * x_test + b

plt.plot(x_test, y_test)
plt.scatter(x_train, y_train)
plt.show()
```
